```python
def retorno_semanal(lista, prob_ganancia, prob_perdida, porc_ganancia, porc_perdida):
  ganancia=0
  perdida=0
  for i in range(len(lista)):
    ganancia=(lista[i]*prob_ganancia*porc_ganancia)+ganancia
    perdida=(lista[i]*prob_perdida*porc_perdida)+perdida
  retorno=ganancia-perdida
  return retorno
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y_line = [2, 4, 6, 8, 10]
y_bar = [1, 3, 5, 7, 9]

# Creating a figure and axis
fig, ax = plt.subplots()

# Plotting the line
ax.plot(x, y_line, label='Line Plot', marker='o', color='blue')

# Plotting the bar
ax.bar(x, y_bar, label='Bar Plot', color='orange', alpha=0.7)

# Adding labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Matplotlib Line and Bar Plot Example')

# Adding a legend
ax.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# Create some sample data for the plots
categories = ['Category A', 'Category B', 'Category C', 'Category D']
bar_data = [25, 50, 75, 100]
pie_data = [30, 20, 25, 25]

# Create a PdfPages object to save the plots to the PDF
pdf_filename = 'bar_and_pie_plots.pdf'

with PdfPages(pdf_filename) as pdf:
    # Bar plot
    plt.figure()
    plt.bar(categories, bar_data, color='skyblue')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.title('Bar Plot')
    pdf.savefig()
    plt.close()

    # Pie plot
    plt.figure()
    plt.pie(pie_data, labels=categories, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightskyblue', 'lightyellow'])
    plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
    plt.title('Pie Plot')
    pdf.savefig()
    plt.close()

print(f'Bar and pie plots saved to {pdf_filename}')
```
